```python
import pandas as pd
```

```python
df = pd.read_csv("customers-100.csv")
```

```python
df.head()
df.info()
df.shape
```

```python
df = df.drop_duplicates()
print("Duplicate rows:", df.duplicated().sum())
```

```python
if "Index" in df.columns:
    df = df.drop(columns=["Index"])
```

```python
print("Missing values before cleaning:")
print(df.isna().sum())

# Fill missing Company with "Unknown"
df["Company"] = df["Company"].fillna("Unknown")

# Fill missing Country with most common value
df["Country"] = df["Country"].fillna(df["Country"].mode()[0])

print("Missing values after cleaning:")
print(df.isna().sum())
```

```python
df["Subscription Date"] = pd.to_datetime(df["Subscription Date"], errors="coerce")
```

```python
df["first_name"] = df["first_name"].str.lower().str.strip()
df["Last Name"] = df["Last Name"].str.lower().str.strip()
df["Country"] = df["Country"].str.lower().str.strip()
df["Email"] = df["Email"].str.strip()
```

```python
print("Final duplicate count:", df.duplicated().sum())
print("Final missing values:")
print(df.isna().sum())
```

```python
df.to_csv("cleaned_customers.csv", index=False)
print("Cleaned file saved successfully!")
```
